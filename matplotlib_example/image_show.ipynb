#### Отображение картинок

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

img = Image.open("речпорт(960×541).jpg")

fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot()

ax.imshow(img)

# plt.show()
```

```python

data = np.random.randint(0, 255, (100, 100))
# data = np.random.randint(0, 255, (96, 96, 3), dtype=np.uint8)
plt.imshow(data)
```

```python
# добавление цветовой карты
data = np.random.randint(0, 255, (100, 100))
plt.imshow(data, cmap="plasma")
```

```python
img = np.array(Image.open("речпорт(960×541).jpg"))

b = plt.imshow(img, origin="lower", cmap="gray", aspect="equal", alpha=0.7)
plt.colorbar(b)
```
