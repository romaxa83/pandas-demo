#### Точечный график

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

x = np.random.normal(0, 2, 500)
y = np.random.normal(0, 2, 500)

ax.scatter(x, y)

ax.grid()
```

Параметры:
 - s - масштаб точек
 - c или color - цвет точек
 - cmap - цветовая схема
 - alpha - степень прозрачности
 - linewidths - толщина вокруг точек
 - edgecolor - цвет границ
 - marker - тип маркера

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

x = np.random.normal(0, 2, 500)
y = np.random.normal(0, 2, 500)

ax.scatter(x, y, s=50, c='g', linewidth=1, marker='s', edgecolor='r')

ax.grid()
```
