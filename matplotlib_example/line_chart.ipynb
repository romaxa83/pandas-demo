#### Заливка областей графика
```python
fill_between(x, y1, y2, where=None, interpolate=False, step=None, *, data=None, **kwds)

# x, y1 - массив значений координат x и функции y1
# y2 - массив(или число) для второй кривой, до которой производится заливка
# where - массив булевых эл., которые определяют область для заливки
```

````python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x,y)
plt.fill_between(x, y)

plt.fill_between(x, y, where=(y < 0), color='red', alpha=0.5)
plt.fill_between(x, y, where=(y > 0), color='yellow', alpha=0.5)
plt.grid(True)
````

#### Несколько координатных осей
для отрисовки нескольких графиков в одной фигуре (холсте), можно использовать функции: `subplot()` или `subplots()`
```python
    sublot(nrows, ncols, index)
    # nrows, ncols - число строк и столбцов
    # index - индек текущих координатных осей
```

```python
    sublots(nrows, ncols)
```

Для более сложной компановки графиков используется объект `GridSpec` из модуля `matplotlib.gridspec`
```python
    gs = GridSpec(ncols=3, nrows=2, figure=fig)
```

````python
ax1 = plt.subplot(1,3,1)
plt.plot(np.random.random(10))
ax2 = plt.subplot(1,3,2)
plt.plot(np.random.random(10))
ax3 = plt.subplot(1,3,3)
plt.plot(np.random.random(10))

ax2.grid(True)
ax3.grid(True)
````

````python
ax1 = plt.subplot(2,2,1)
plt.plot(np.random.random(10))
ax2 = plt.subplot(2,2,2)
plt.plot(np.random.random(10))
ax3 = plt.subplot(2,1,2)
plt.plot(np.random.random(10))

ax2.grid(True)
ax3.grid(True)
````

````python
figure, ax = plt.subplots(2,2)

# https://matplotlib.org/stable/api/figure_api.html
figure.set_size_inches(7,4)
figure.set_facecolor('#eee')

ax[0,0].plot(np.random.random(10))
ax[0,1].grid(True)
ax[1,1].plot(np.arange(0, 5, 0.5))
````

````python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(7,4))
gs = GridSpec(ncols=3, nrows=2, figure=fig)

ax1 = plt.subplot(gs[0,0])
plt.plot(np.arange(0, 5, 0.5))

ax2 = plt.subplot(gs[1,0:2])
plt.plot(np.random.random(10))

# можно и так добавлять координатные оси
ax3 = fig.add_subplot(gs[:,2])
plt.plot(np.random.random(30))
````

````python
# можно задавать размер осям

# Размер по ширине указывает какую долю по ширине будут занимать столбцы, т.е. ширина первого столбца 1/8, второго - 2/8, третьего - 5/8
ws = [1, 2, 5]
# размер по высоте
hs = [2, 0.5]

fig = plt.figure(figsize=(7,4))
gs = GridSpec(ncols=3, nrows=2, figure=fig, width_ratios=ws, height_ratios=hs)

ax1 = plt.subplot(gs[0,0])
plt.plot(np.arange(0, 5, 0.5))

ax2 = plt.subplot(gs[1,0:2])
plt.plot(np.random.random(10))

ax3 = fig.add_subplot(gs[:,2])
plt.plot(np.random.random(30))
````

#### Положение меток на координатных осях
- `set_major_locator()` - управление рисками крупной сетки
- `set_minor_locator()` - управление рисками мелкой сетки

В них нужно передать локатор, для манипулирования метками

````python
from matplotlib.ticker import (
    NullLocator,
    LinearLocator,
    MultipleLocator,
    IndexLocator,
    FixedLocator,
    LogLocator,
    MaxNLocator, FixedFormatter
)

fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.5))

ax.grid()
# вызываем для объекта оси x
ax.xaxis.set_major_locator(NullLocator())
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.5))

ax.grid()
# вызываем для объекта оси x
ax.xaxis.set_major_locator(LinearLocator(3))
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.25))

ax.grid()
# base - шаг для рисок
ax.xaxis.set_major_locator(MultipleLocator(base=1))
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.25))

ax.grid()
# данный локатор работает MultipleLocator, но можно указать offset - откуда начинаются риски
ax.xaxis.set_major_locator(IndexLocator(base=1, offset=8))
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.25))

ax.grid()
# данный локатор задает фиксированные риски
ax.xaxis.set_major_locator(FixedLocator([-2,-1,0,1,2,3, 10]))
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.25))

ax.grid()
# данный локатор использует логарифмическое деление по осям
ax.xaxis.set_major_locator(LogLocator(2))
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.25))

ax.grid()
# данный локатор использует максимальное кол-во указанных рисок
#  https://matplotlib.org/api/ticker_api.html#matplotlib.ticker.MaxNLocator
ax.xaxis.set_major_locator(MaxNLocator(6))
````

````python
# использование минорной сетки

fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.25))

ax.minorticks_on()
ax.grid(which='major', lw=2)
ax.grid(which='minor')

ax.xaxis.set_minor_locator(MaxNLocator(6))
````

#### Формат отображения меток
с помощью функций можно отключить отображение меток:
- `set_xticklabels()`
- `set_yticklabels()`

для добавления собственных меток нужно использовать форматеры через функцию
- `set_major_formatter()`

````python
from matplotlib.ticker import (
    NullFormatter,
    FormatStrFormatter,
    FuncFormatter,
    ScalarFormatter,
    FixedFormatter,
)

#https://matplotlib.org/stable/api/ticker_api.html

fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
plt.plot(np.arange(0, 5, 0.25))

# отключаем метки по оси x
ax.set_xticklabels([])
# отключаем метки по оси y
ax.yaxis.set_major_formatter(NullFormatter())
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

# отображает только целые числа
# ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
# отображает только вещественные значения (округляем до тысячных)
# ax.yaxis.set_major_formatter(FormatStrFormatter('%.3f'))
ax.yaxis.set_major_formatter(FormatStrFormatter('y = %.3f'))
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

def format0y(x, pos):
    return f"[{x}]" if x < 0 else f"({x})"

# вызывает функцию для формирования числовых значений
ax.yaxis.set_major_formatter(FuncFormatter(format0y))
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x) * 1e5)

# данный форматер используется по умолчанию
sf = ScalarFormatter()
# указываем после какой степени выводить ее за скобки
sf.set_powerlimits((-4, 4))
# sf.set_powerlimits((-6, 6))

# так можно указать глобально для всех графиков
# matplotlib.rcParams['axes.formatter.limits'] = (-4,4)

ax.yaxis.set_major_formatter(sf)
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

# каждой риски присваивается свое значение
ax.xaxis.set_major_formatter(FixedFormatter(['a', 'b', 'c', 'd', 'e', 'f', 'g']))
ax.grid()
````

#### Добавление текстовых поле к осям координат

У каждой оси (объекта Axes) можно определять следующие текстовые элементы:
- title - заголовок осей
- `xlabel`, `ylabel` - подписи для каждой оси
- text - произвольная текстовая информация в области осей
- annotate - аннотация (текст с указателем)

Также у окна в целом можно задавать такие текстовые элементы:
- subtitle - заголовок для фигуры (окна)
- figtext - размещение текстовой информации в области окна

````python
# Свойства для текстовых элементов: https://matplotlib.org/stable/api/text_api.html

fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
# первые два параметра - координаты текста, диапазон от 0 до 1
plt.figtext(0.05, 0.6, "Текст в области окна")
fig.suptitle('Заголовок окна')
ax.set_xlabel('0x')
ax.set_ylabel('0y')
# plt.xlabel('0x')
# plt.ylabel('0y')
ax.text(0.05, 0.1, 'Текст в координатных осях')
ax.annotate('Аннотация', xy=(0.2, 0.4), xytext=(0.6, 0.7), arrowprops={'facecolor':'red', 'shrink':0.1})
````

````python
# можем создавать рамки вокруг текста
fig = plt.figure(figsize=(7,4), facecolor='red')
ax = fig.add_subplot()
ax.set(facecolor='#AAFFAA')
# первые два параметра - координаты текста, диапазон от 0 до 1
plt.figtext(0.05, 0.6, "Текст в области окна")

ax.text(0.05, 0.1, 'Текст в координатных осях', bbox={'boxstyle':'darrow', 'alpha':0.5, 'facecolor':'#AAAAFF'})

# Параметры bbox:
# - boxstyle - вид рамки вокруг текста
# - alpha - степень прозрачности фона
# - color - цвет фона с рамкой
# - ec - цвет рамки
# - facecolor - цвет заливки
# - fill - использовать ли заливку (True/False)
# - hatch - тип штриховки ('/', '\', '|', '-', '+', 'x', 'o', 'O', '.', '*')
# - linestyle - стиль линии границ ('-', '--', '-.', ':')
# - linewidth - толщина рамки

# Аттрибуты boxstyle
# - cicrle, darrow, larrow, rarrow, round, roundtooth, square
````

#### Логарифмический масштаб для координатных осей

````python
# используется когда данных очень много и линейный график не отображает всех изменений
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
x = np.arange(-10*np.pi, 10*np.pi, 0.5)

# если так вызвать, то по краям графика не заметны изменения
ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10) ))
# если добавить по оси y логарифмический масштаб, то видны изменения по краям
ax.set_yscale('log')

# или можно так вызвать
# ax.semilogy(x, np.sin(x) * np.exp( -np.abs(x/10) ))

ax.grid()
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
x = np.arange(-10*np.pi, 10*np.pi, 0.5)

ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10) ))
# base - по какому основанию применяется логарифмическая функция, subs - какие риски, между пределами отображать
ax.set_yscale('log', base=10, subs=[2,9])

ax.grid()
````

````python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
x = np.arange(-10*np.pi, 10*np.pi, 0.1)

ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10) ))
# параметр symlog - указывает что будет использоваться и линейный и логарифмический масштаб
ax.set_xscale('symlog', linthresh=2)
# ax.set_xscale('symlog', linthresh=2, linscale=5)

ax.grid()
````
