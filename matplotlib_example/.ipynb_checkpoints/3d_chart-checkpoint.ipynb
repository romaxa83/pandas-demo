#### Трехмерные графики

```python
# https://matplotlib.org/stable/api/_as_gen/mpl_toolkits.mplot3d.axes3d.Axes3D.html
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(7, 4))
ax_3d = fig.add_subplot(projection='3d')

x = np.linspace(0, 10, 50)
z = np.cos(x)
ax_3d.plot(x, x, z)

ax_3d.set_xlabel('x')
ax_3d.set_ylabel('y')
ax_3d.set_zlabel('z')

# plt.show()
```

```python
# plot - линейный 2D график в трех измерениях
# step - ступенчатый 2D график в трех измерениях
# scatter - точечный 3D график
# plot_wireframe - построение каркасной поверхности в 3D
# plot_surface - построение непрерывной 3D поверхности
```

```python
# строим синусойду в трехмерном графике
fig = plt.figure(figsize=(7, 4))
ax_3d = fig.add_subplot(projection='3d')

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.arange(-2*np.pi, 2*np.pi, 0.01)

xgrid, ygrid = np.meshgrid(x,y)

zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

# ax_3d.plot_wireframe(xgrid, ygrid, zgrid)
# ax_3d.plot_surface(xgrid, ygrid, zgrid)
# ax_3d.plot_surface(xgrid, ygrid, zgrid, rstride=5, cstride=5, cmap='plasma')
ax_3d.scatter(xgrid, ygrid, zgrid, s=1, color='g')
```

#### Рисуем линии контура

Используя функции
- `contour()` и `contourf()` - если данные по осям x, y, z представлены в виде двумерных массивов
- `tricontour()` и `tricontourf()` - если данные по осям x, y, z представлены в виде одномерного массивов

```python
fig = plt.figure(figsize=(7, 4))
ax_3d = fig.add_subplot(projection='3d')

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.arange(-2*np.pi, 2*np.pi, 0.01)

xgrid, ygrid = np.meshgrid(x,y)

zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

ax_3d.contour(xgrid, ygrid, zgrid)

# линии контура в виде поверхностей
# ax_3d.contourf(xgrid, ygrid, zgrid)
```

```python
# линии контура в двухмерном графике
fig, ax = plt.subplots(1, 2)

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.arange(-2*np.pi, 2*np.pi, 0.01)

xgrid, ygrid = np.meshgrid(x,y)

zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

ax[0].contour(xgrid, ygrid, zgrid)
ax[1].contourf(xgrid, ygrid, zgrid)
```

```python
# выводим значения линий контура в двухмерном графике
fig, ax = plt.subplots(1,2)

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.arange(-2*np.pi, 2*np.pi, 0.01)

xgrid, ygrid = np.meshgrid(x,y)

zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

c1 = ax[0].contour(xgrid, ygrid, zgrid, colors=['g', 'b', 'r'])
c2 = ax[1].contour(xgrid, ygrid, zgrid, 15)

ax[0].clabel(c1)
ax[1].clabel(c2, colors='r', fmt='%.4f')
# clabel(): https://matplotlib.org/stable/api/contour_api.html#matplotlib.contour.ContourLabeler.clabel
```

```python
# выводим значения линий контура в двухмерном графике
fig, ax = plt.subplots(1,2)

x = np.random.rand(100) * 4*np.pi - 2*np.pi
y = np.random.rand(100) * 4*np.pi - 2*np.pi
z = np.sin(x) * np.sin(y) / (1+np.abs(x*y))

c1 = ax[0].tricontour(x, y, z)
c2 = ax[1].tricontourf(x, y, z)

ax[0].clabel(c1)
ax[1].clabel(c2, colors='r', fmt='%.4f')
```
