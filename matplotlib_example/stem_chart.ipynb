#### Стем график

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

# строим синусойду
x = np.arange(-np.pi, np.pi, 0.3)
ax.stem(x, np.cos(x))

ax.grid()
```

```python
# меняем базовый уровень
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

x = np.arange(-np.pi, np.pi, 0.3)
ax.stem(x, np.cos(x), '--', bottom=0.5)

ax.grid()
```
