#### Гистограмма

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

x = np.random.normal(0, 2, 500)

# по умолчанию распределит группы значений на 10 столбцов
# ax.hist(x)

ax.hist(x, 50)

ax.grid()
```

```python
# можно самому выстроить через функцию bar

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(1,5, len(x))

ax.bar(x,y)
ax.grid()
```

```python
# вертикальное отображение

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(1,5, len(x))

ax.barh(x,y)
ax.grid()
```

```python
# добавление стилей и уровня

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(-20, 20, len(x))

ax.bar(x,y, width=0.5, linewidth=2, edgecolor='r', yerr=2, bottom=10)
ax.grid()
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

x = np.arange(10)
y1 = np.random.randint(3, 20, len(x))
y2 = np.random.randint(3, 20, len(x))
w = 0.3

ax.bar(x - w/2, y1, width=w)
ax.bar(x + w/2, y2, width=w)

ax.grid()
```
