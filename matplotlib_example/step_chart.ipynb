#### Ступенчатый (step) график

часто в науки используется для отображения цифровых данных, когда каждое значение имеет строго определенный уровень

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

x = np.arange(0,10)
ax.step(x,x, '-o')

# инверсия дефолтного значение, которое имеет значение 'pre'
# ax.step(x, x, '-o', where='post')

# точки будут по центру
# ax.step(x,x, '-o',where='mid')
ax.grid()
```

```python
# два графика
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot()

x = np.arange(0,10)
ax.step(x,x, '-o', x, np.cos(x), '--x', where='mid')

ax.grid()
```
