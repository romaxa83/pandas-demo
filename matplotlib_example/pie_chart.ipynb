#### Круговые диаграммы

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

vals = [10, 40, 23, 30, 70]
labels = ["Toyota", "BWM", "Lexus", "Audi", "Tavria"]

ax.pie(vals, labels=labels)

ax.grid()
```

Параметры:

- labels - список подписей для долей
- explode - список долей, выносимых их диаграммы
- colors - цвет долей
- autopct - формат числа величины доли внутри сегмента
- pctdistance - расстояние от центра доли до текстовой метки
- shadow - отображение тени
- labeldistance - расстояние для текстовых полей (по умолчанию (1,1))
- startangle - начальный угол поворота диаграммы
- radius - радиус
- counterclock - порядок размещение долей на диаграмме, по часовой/против часовой (True/False)
- center - координата центра диаграммы (по умолчанию (0,0))
- frame - отображение рамки вокруг диаграммы (True/False)
- wedgeprops - словарь доп. параметров

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

vals = [10, 40, 23, 30, 70]
labels = ["Toyota", "BWM", "Lexus", "Audi", "Tavria"]
exp = (0.1, 0.2, 0, 0, 0)
ax.pie(vals, labels=labels, autopct='%.2f', explode=exp, shadow=True)

ax.grid()
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()

vals = [10, 40, 23, 30, 70]
labels = ["Toyota", "BWM", "Lexus", "Audi", "Tavria"]
exp = (0.1, 0.2, 0, 0, 0)
ax.pie(vals, labels=labels, autopct='%.2f', wedgeprops=dict(width=0.5), shadow=True)

ax.grid()
```
